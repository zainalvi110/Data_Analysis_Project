```python
import pandas as pd
dataframe=pd.read_csv('orders.csv',na_values=['Not Available','unknown'])
dataframe.head(15)
```

```python
dataframe['Ship Mode'].unique()
```

```python
dataframe.columns = dataframe.columns.str.lower()
dataframe.columns=dataframe.columns.str.replace(' ','_')
```

```python
dataframe.columns
```

```python
dataframe.head()
```

```python
# making new column
dataframe['list_price']*dataframe['discount_percent']*.01
```

```python
dataframe['discount']=dataframe['list_price']*dataframe['discount_percent']*.01
```

```python
dataframe.head()
```

```python
dataframe['sale_price']=dataframe['list_price'] - dataframe['discount']
```

```python
dataframe.head()
```

```python
dataframe['profit']= dataframe['sale_price'] - dataframe['cost_price']
```

```python
dataframe.head()
```

```python
dataframe['order_date']=pd.to_datetime(dataframe['order_date'],format="%Y-%m-%d")
```

```python
dataframe.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
```
